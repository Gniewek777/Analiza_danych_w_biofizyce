# Wstęp teoretyczny

Na wykladzie omawialiśmy instrukcje sterujące i funkcje. Z instrukcji sterujacych omówiliśmy:  
1. Instrukcje sterującą if...elif..else  
2. Pętle:
    * while
    * for

Pętla `for` definiujemy w taki sposób `for i in kontener` gdzie i jest obiektem kontenera i można na nim operować.  
Pętla `while` wykouję jakąś czynność dopóki warunek wewnątrz niej nie jest spełnony. Definiujemy w taki sposob: 
`while warunek: #coś zrób`

W pythonie mamy sowa klucze które pozwalaja modyfikować pętle a są nimi:
* `break`- natychmiastowo przerywa działanie kodu,
* `pass`- nic nie robi 
* `continue` - wymusza przejscie do nastepnej iteracji
* `else` - wykonuje instrukcje wprzypadku gdy nie doszło do wymuszenia przerwania petli instrukcja break

Póżniej przeszliśmy do funkcji. Dzieli się ona na 3 bloki:
1. lista parametrów - wszystkieo biekty przekazywane
2. ciało funkcji - zestaw instrukcji do wykonania
3. zwracane obiekty - wyniki.

Warto pamiętać, że przy definiowaniu funkcji robimy wcięcie. Definiujemy ją w ten sposób: `def funkcja(x,y): #coś`
W funkcji mogą wystepować argumenty typu **klucz = wartość** (`funkcja(x,y=1)`) ale należy pamiętać,  że argumenty tego typu przekazujemy na końcu listy argumentów funkcji, ponieważ argumenty dopasowane pozycyjnie(czyli kolejno) są ropatrywane jako pierwsze.Można również w funkcji implementowac wartości domyślne `def funkcja(x="coś",y):`.

Kolejną klasą parametrów są parametry specjalne:
1. *args - grupuje "dodatkowe" argumenty pozycyjne (krotka).
2. **kwargs - pozwala na zbiorczą reprezentację szystkich niewymienionych na liscie parametrów argumentów hasłowych (słownik).

Na koniec poznaliśmy możliwość zapisywania wszystkich zdefiniowanych funkcji w jednym pliku a następnie odwoływac się do nich za pomoca `import`.

# Zadanie 4 - Całkowanie sygnałów - lepsza aproksymacja (level up)

```python
def calkuj2(f,a,b):
    h =(b-a)/2
    return ((f(a)+f(h))/2)*h + ((f(h)+f(b))/2)*h
```

```python
print(calkuj2(f1,0,log(3)))
print(calkuj2(f2,0,pi))
print(calkuj2(f3,0,pi))
```

Policzone zostały całki metodą numeryczną zwaną metodą trapezoidów z użyciem dwóch trapezów.


# Zadanie 5 - Całkowanie sygnałów - czy może być lepiej?


```python
def calkuj3(f,a,b,n):
        dx = (b-a)/n
        pole = [f(a+i*dx) for i in range(1,n)]
        return dx*(sum(pole)+(f(a)+f(b))/2)
```

```python
print(calkuj3(f1,0,log(3),3))
print(calkuj3(f2,0,pi,3))
print(calkuj3(f3,0,pi,3))

print(calkuj3(f1,0,log(3),10))
print(calkuj3(f2,0,pi,10))
print(calkuj3(f3,0,pi,10))
```

Policzone zostały całki metodą numeryczną zwaną metodą trapezoidów z użyciem n trapezów kolejno dla n = 3 i n = 10.
