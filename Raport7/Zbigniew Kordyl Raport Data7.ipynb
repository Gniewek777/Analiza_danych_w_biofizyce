# Wstęp teoretyczny

Na wykładzie kontynuowaliśmy temat wektorów i macierzy z biblioteki numpy. Tam poznaliśmy kilka przydatnych metod:  

* reshape(w1,...) - Modyfikuje kształt tabeli o unikalne zgodne wymiary.
* ravel() - Dokonuje likwidacji wirtualnego podziału tabelina wymiary (zwraca wektor).
* arange(a,b,l) - Tworzy ciag liczb zawartych w przedziale [a,b)] i przyrostach równych l.
* eye(n) - Tworzy macierz jednostkowa o wymiarze n×n.
* diag([a,b..])- Tworzy macierz diagonalna o elementachdiagonalnycha,b, ...
* zeros(tuple)) - Tworzy macierz wypełniona zerami.
* append(t1, t2) - Dodaje do istniejacej tabeli nowe elementy.
* vstack(t1, t2..) - Łaczy wierszowo krotke obiektów tablicopochodnych.
* hstack(t1, t2..) - Łaczy kolumnowo krotke obiektów tablicopochodnych.
* column_stack(t1,..) - Łaczy kolumnami wektory i macierze.

Wszystkie operacje jak operatory binarne,relacyjne i logiczne na macierzach z biblioteki numpy sa zwektoryzowane przez co ułatwia to nam bardzo prace i niektóre obliczenia przez co nie musimy iterowac po każdym elemencie macierzy żeby na przyklad coś dodać. Należy jednka pamietac żeby korzystac z metod biblioteki numpy, ponieważ wtedy wektoryzacja nie zadziała.  

Należy również pamietać żeby macierze były o zgodnych krztałtach, ponieważ może to utrudnić pewne operacje.

Biblioteka numpy oferuje też przydatne metody zwektoryzowane do agregacji danych jak:
* sum() - suma
* prod() - iloczyn
* mean() - srednia arytmetyczna
* std() - odchylenie standardowe
* var() - wariancja
* max() - element największy
* min() - element najmniejszy

Z obiektów array można wybierać podzbiory danych w celu np. filtrowania.

````python
g = 980.665 # [cm/s^2]  wartość ziemskiego przyspieszenia grawitacyjnego
s = 7.9 * 10**(-4) # [N/cm] współczynnik napięcia powierzchniowego układu powietrze-woda
ρ = 1 # [g/cm3] gęstość wody
h = 500 # [cm] wysokość słupa wody (głębokość)

def c(lam):
    return np.sqrt((g*lam)/(2*np.pi) * (1 + s * ((4*np.pi**2)/(ρ*g*lam**2) * np.tanh((2*np.pi*h)/(lam)))))

lam_1 = np.linspace(0.1,10,100)
lam_2 = np.linspace(100,200000,100)

fig, axs = plt.subplots(2)
fig.suptitle('Prędkość czoła fali v wzbudzonej na powierzchni wody zależnej od długości  fali λ')

axs[0].plot(lam_1, c(lam_1))
axs[1].plot(lam_2, c(lam_2))


for ax in axs.flat:
    ax.set(xlabel='Długość fali λ [cm]', ylabel='Prędkość czoła fali ')
````

````python
def Lk(x,k,punkty):
    res = 1
    for i in range(0,len(punkty[0])):
        if i == k:
            continue
        res *=(x-punkty[0][i])/(punkty[0][k]-punkty[0][i])
    return res

def Lagrange(x,punkty):
    suma = 0
    for k in range(0,len(punkty[0])):
        suma += punkty[1][k]*Lk(x,k,punkty)
    return suma

def sprawdzam(punkty):
    s =0
    for x,y in zip(punkty[0],punkty[1]):
        s+=abs(Lagrange(x,punkty)-y)
    return s==0
````

````python
x = np.linspace(0,2*np.pi,10)
y = np.sin(x)
s = np.array((x,y))
sprawdzam(s)
````
