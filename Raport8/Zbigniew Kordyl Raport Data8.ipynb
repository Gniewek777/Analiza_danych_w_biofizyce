# Wstęp teoretyczny

Na wykładzie zaczeliśmy temat ramek danych. Są to to macierze w których każda kolumna w stanie przehcowywać dane różnych typów. A by z nich kozystać musimy zaimportować biblioteke pandas ``` import pandas as pd ```. Następnie aby skonstruować taką ramke można użyć słownika bądź macierzy (kostruktor) jako dane wejściowe. 
Przykład:
``` python
dane = {"Pacjent": ["Tomek", "Mateusz"],"Wiek":[20,11],"Krew":["A","B"]}
pac_df = pd.DataFrame(dane)
print(pac_df)

  Pacjent  Wiek Krew
0    Tomek    20    A
1  Mateusz    11    B

```
W przypadku macierzy nazwy kolumn nalezy wygenerowac recznie, wykorzystując argument ```columns``` tego konstruktora.

Poznalismy podstawowe cechy ramek:
* kształt ramki- zawarte w polushape,które przechowuje informacj ̨e o liczbiewierszy i kolumn danej ramki.
* liczba elementów- zawarte w polusize,które przechowuje informacj ̨e o liczbieelementów (komórek) danej ramki.
* iczba instancji danych/obserwacji-zwraca je metodalen()
* typy danych- zawarte w poludtypes
* podstawowe statystyki- zwraca je metodainfo()klasyDataFrame.

Ramki danych posiłkują się z notacji słowika przez co aby dostać się do informacji z podanej kolumny wystraczy odwołać się w taki sposób: ```ramka["zmienna"]```, ale można też na dwa inne sposoby: ```ramka.get("zmienna")``` i ```ramka.zmienna```.

Aby odwoływać się do etykiet  wierszy ramki należy uzyć komendy ```ramka.index``` a w przypadku kolumn ```ramka.columns```

Jeśli chcemy ręcznie modyfikować etykiety możemy zrobić to w następujący sposób: ```ramka.set_index(nowe_etykiety)```  oraz  
nadpisując pola index/columns ```ramka.index = [....]```, ```ramka.columns = [....]```

Poznaliśmy przydatne metody jak:
* ramka.tail() która zwraca "skrócona" ramke danych, zawierajaca N=5 ostatnich instancji danych
* ramka.head() która zwraca "skrócona" ramke danych, zawierajaca N=5 pierwszych instancji danych

Możemy takie ramki indeksować za pomocą pól:
* iloc() gdzie możemy odnosić się do elementów ramki podobnie jak w macierzy ndarray np. ramka.iloc[1:−1]
* loc() gdzie mamy możliwość dostępu do etykiet np. pac_data.loc[:3, ’serum_sodium’:’smoking’]
